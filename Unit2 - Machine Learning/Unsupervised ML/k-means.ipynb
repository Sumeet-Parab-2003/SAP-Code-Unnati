```python
import matplotlib.pyplot as plt
import seaborn as sns


# Set the aesthetic style of the plots
sns.set_style('whitegrid')

# Plot distributions of numerical features
plt.figure(figsize=(20, 5))

# Distribution of Age
plt.subplot(1, 3, 1)
sns.histplot(df['Age'], bins=30, kde=True)
plt.title('Age Distribution')

# Distribution of Annual Income
plt.subplot(1, 3, 2)
sns.histplot(df['Annual Income (k$)'], bins=30, kde=True)
plt.title('Annual Income Distribution')

# Distribution of Spending Score
plt.subplot(1, 3, 3)
sns.histplot(df['Spending Score (1-100)'], bins=30, kde=True)
plt.title('Spending Score Distribution')

plt.tight_layout()
plt.show()
```

```python
# Group the data by age and calculate the average spending score for each age
df_age_spending = df.groupby('Age')['Spending Score (1-100)'].mean().reset_index()

# Find the age group with the highest average spending score
max_spending_age = df_age_spending[df_age_spending['Spending Score (1-100)'] == df_age_spending['Spending Score (1-100)'].max()]

# Visualize the relationship between age and average spending score
plt.figure(figsize=(10, 6))
sns.barplot(x='Age', y='Spending Score (1-100)', data=df_age_spending.sort_values('Spending Score (1-100)', ascending=False))
plt.title('Average Spending Score by Age')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()

max_spending_age
```

```python
# Calculate the spending score to income ratio for each customer
df['Spending_to_Income_Ratio'] = df['Spending Score (1-100)'] / df['Annual Income (k$)']

# Group the data by age and calculate the average ratio for each age group
df_age_income_ratio = df.groupby('Age')['Spending_to_Income_Ratio'].mean().reset_index()

# Find the age group with the highest average ratio
max_ratio_age = df_age_income_ratio[df_age_income_ratio['Spending_to_Income_Ratio'] == df_age_income_ratio['Spending_to_Income_Ratio'].max()]

# Visualize the relationship between age and spending to income ratio
plt.figure(figsize=(10, 6))
sns.barplot(x='Age', y='Spending_to_Income_Ratio', data=df_age_income_ratio.sort_values('Spending_to_Income_Ratio', ascending=False))
plt.title('Average Spending to Income Ratio by Age')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()

max_ratio_age
```

```python
# Filter the dataframe for the age group 18 to 40
df_age_18_40 = df[(df['Age'] >= 18) & (df['Age'] <= 40)]

# Get a high-level overview of the dataframe for age group 18 to 40
df_age_18_40.describe(include='all')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the aesthetic style of the plots
sns.set_style('whitegrid')

# Plotting the distribution of Annual Income based on Gender for age group 18-40
plt.figure(figsize=(14, 6))
sns.histplot(data=df_age_18_40, x='Annual Income (k$)', hue='Gender', kde=True, element='step', bins=15)
plt.title('Distribution of Annual Income based on Gender for Age 18-40')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Count')
plt.legend(title='Gender')
plt.show()

# Plotting the distribution of Spending Score based on Gender for age group 18-40
plt.figure(figsize=(14, 6))
sns.histplot(data=df_age_18_40, x='Spending Score (1-100)', hue='Gender', kde=True, element='step', bins=15)
plt.title('Distribution of Spending Score based on Gender for Age 18-40')
plt.xlabel('Spending Score (1-100)')
plt.ylabel('Count')
plt.legend(title='Gender')
plt.show()

# Summary statistics based on Gender
summary_by_gender = df_age_18_40.groupby('Gender').agg({'Annual Income (k$)': ['mean', 'std', 'min', 'max'], 'Spending Score (1-100)': ['mean', 'std', 'min', 'max']})
summary_by_gender
```

```python
# Visualize the summary statistics for Annual Income and Spending Score by Gender
plt.figure(figsize=(14, 7))

# Create a bar plot for average annual income by gender
plt.subplot(1, 2, 1)
sns.barplot(x=summary_by_gender.index, y=summary_by_gender['Annual Income (k$)']['mean'], yerr=summary_by_gender['Annual Income (k$)']['std'], palette='pastel')
plt.title('Average Annual Income by Gender')
plt.xlabel('Gender')
plt.ylabel('Average Income (k$)')

# Create a bar plot for average spending score by gender
plt.subplot(1, 2, 2)
sns.barplot(x=summary_by_gender.index, y=summary_by_gender['Spending Score (1-100)']['mean'], yerr=summary_by_gender['Spending Score (1-100)']['std'], palette='pastel')
plt.title('Average Spending Score by Gender')
plt.xlabel('Gender')
plt.ylabel('Average Spending Score')

plt.tight_layout()
plt.show()
```

```python
from sklearn.cluster import KMeans 
```

```python
def find_best_clusters(df,maximum_k): 
    centroids = []
    k_values = []
    for k in range (2 , maximum_k):
        kmeans = KMeans(n_clusters=k)
        kmeans.fit(df)
        
        centroids.append(kmeans.inertia_)
        k_values.append(k)
    return centroids,k_values
```

```python
centroids,k_values= find_best_clusters(scaled_data,12)
```
