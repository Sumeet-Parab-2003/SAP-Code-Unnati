```python
import seaborn as sns
```

```python
# Box plot for income based on gender
plt.figure(figsize=(8, 6))
sns.boxplot(x='gender', y='income', data=df)
plt.title('Income Distribution by Gender')
plt.xlabel('Gender')
plt.ylabel('Income')
plt.show()
```

# Income Distribution by gender

```python
# Analyze based on income
plt.figure(figsize=(12, 6))
sns.histplot(df['income'], bins=30, kde=True)
plt.title('Distribution of Income')
plt.xlabel('Income')
plt.ylabel('Count')
plt.show()
```

# As the salary increases Count of employees having high salay decreases

```python
# Countplot for gender based on region
plt.figure(figsize=(10, 6))
sns.countplot(x='region', hue='gender', data=df)
plt.title('Count of Customers by Gender in Each Region')
plt.xlabel('Region')
plt.ylabel('Count')
plt.show()
```
