```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import warnings 
warnings.filterwarnings("ignore")
%matplotlib.incline
```

```python
x = np.arange(0,10)
y = np.arange(40,50)
```

```python
y=x*x
plt.plot()
plt.plot(x,y,"*-",linestyle = "dashed",linewidth = 3 , markersize=10)
plt.xlabel('x')
plt.ylabel('y')
plt.title("2-d plot")
```
